# Purchasing Analysis (Gender)

    ■ Purchase Count

    ■ Average Purchase Price
    
    ■ Total Purchase Value
    
    ■ Average Purchase Total per Person by Gender

● Create a summary data frame to hold the results

● Display the summary data frame

```python
#分類
male = purchase_data[purchase_data['Gender'] == 'Male']
female = purchase_data[purchase_data['Gender'] == 'Female']
other = purchase_data[purchase_data['Gender'] == 'Other / Non-Disclosed']

#計算average purchase price
average_m =  '%6.5f' % ((male['Price'].sum())/male.shape[0])
average_f =  '%6.5f' % ((female['Price'].sum())/female.shape[0])
average_o =  '%6.5f' % ((other['Price'].sum())/other.shape[0])

#計算真正的男女人數
male2=[]
female2=[]
other2=[]
for i in range(purchase_data.shape[0]):
    if purchase_data['Gender'][i] == 'Male':
        male2.append(purchase_data['SN'][i])
    if purchase_data['Gender'][i] == 'Female':
        female2.append(purchase_data['SN'][i])
    if purchase_data['Gender'][i] == 'Other / Non-Disclosed':
        other2.append(purchase_data['SN'][i])
male2 = len(set(male2))
female2 = len(set(female2))
other2 = len(set(other2))


#計算人均花費
avg_total_m = '$''%3.2f' % ((male['Price'].sum())/male2)
avg_total_f = '$''%3.2f' % ((female['Price'].sum())/female2)
avg_total_o = '$''%3.2f' % ((other['Price'].sum())/other2)



answer = pd.DataFrame({'Purchase Count':[female.shape[0],male.shape[0],other.shape[0]],
                       'Average Purchase Price':[average_f,average_m,average_o],
             'Total Purchase Value':[female['Price'].sum(),male['Price'].sum(),other['Price'].sum()],
             'Avg Purchase Total per Person':[avg_total_f,avg_total_m,avg_total_o]},index=['Female','Male','Other / Non-Disclosed'])
answer.index.name = 'Gender'
answer
```

● Categorize players using the age groups. Hint: use pd.cut()

    ■ each group is for 4 years (i.e. <10, 10-14, 15-19, etc.)
    
● Calculate the numbers and percentages by age group

● Create a summary data frame to hold the results

```python
#移除重複的玩家
players = purchase_data.drop_duplicates(subset=['SN'])

#計算區間人數
age_intervel = pd.value_counts(pd.cut(players['Age'],[0,9,14,19,24,29,34,39,players['Age'].max()+1])).sort_index()

#計算百分比
percentage = ((age_intervel/players.shape[0])*100).round(2)

#印出結果
age_demographics = pd.DataFrame({'Percentage of players':percentage,
                                'Total Count':age_intervel})
age_demographics.index='<10','10-14','15-19','20-24','25-29','30-34','35-39','40+'
age_demographics
```

# Purchasing Analysis (Age)

● The below each broken into groups of 4 years (i.e. <10, 10-14, 15-19, etc.)

    ■ Purchase Count
  
    ■ Average Purchase Price
  
    ■ Total Purchase Value
  
    ■ Average Purchase Total per Person by Age Group
  
● Bin the purchase_data data frame by age

● Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per
person etc. in the table below

● Create a summary data frame to hold the results

● Display the summary data frame

```python
#依Purchase Id來計算區間人數
age_intervel = pd.cut(purchase_data['Age'],[0,9,14,19,24,29,34,39,players['Age'].max()+1])
count = pd.value_counts(age_intervel).sort_index()
count.index = [0,1,2,3,4,5,6,7]

#分類
price1 = purchase_data[purchase_data['Age']<10]['Price']
price2 = purchase_data[(purchase_data['Age']>=10) & (purchase_data['Age']<15)]['Price']
price3 = purchase_data[(purchase_data['Age']>=15) & (purchase_data['Age']<20)]['Price']
price4 = purchase_data[(purchase_data['Age']>=20) & (purchase_data['Age']<25)]['Price']
price5 = purchase_data[(purchase_data['Age']>=25) & (purchase_data['Age']<30)]['Price']
price6 = purchase_data[(purchase_data['Age']>=30) & (purchase_data['Age']<35)]['Price']
price7 = purchase_data[(purchase_data['Age']>=35) & (purchase_data['Age']<40)]['Price']
price8 = purchase_data[purchase_data['Age']>=40]['Price']

#計算
average_price = pd.Series([price1.mean(),price2.mean(),price3.mean(),price4.mean(),
                                   price5.mean(),price6.mean(),price7.mean(),price8.mean()]).round(2)
total_price = pd.Series([price1.sum(),price2.sum(),price3.sum(),price4.sum(),
                price5.sum(),price6.sum(),price7.sum(),price8.sum()]).round(2)

#移除重複的玩家
players = purchase_data.drop_duplicates(subset=['SN'])

#計算區間人數(去除重複玩家後的)
age_intervel = pd.value_counts(pd.cut(players['Age'],[0,9,14,19,24,29,34,39,players['Age'].max()+1])).sort_index()
avg=[]
for i in range(age_demographics.shape[0]):
    avg.append('$'+str((total_price[i]/age_intervel[i]).round(2)))
summary = pd.DataFrame({'Purchase Count':count,
                       'Average Purchase Price':average_price,
                       'Total Purchase Value':total_price,
                       'Average Purchase Total per Persor':pd.Series(avg)})
#在前面加$字符號
summary['Average Purchase Price'] = '$'+summary['Average Purchase Price'].astype(str)
summary['Total Purchase Value'] = '$'+summary['Total Purchase Value'].astype(str)
summary
```

# Top Spenders


● Identify the the top 5 spenders in the game by total purchase value (top-5 players with
highest total purchase value), then list (in a table):
    
    ■ SN(screen name)
    
    ■ Purchase Count
    
    ■ Average Purchase Price
    
    ■ Total Purchase Value
● Run basic calculations to obtain the results in the table below

● Create a summary data frame to hold the results

● Sort the total purchase value column in descending order

● Display a preview of the summary data frame

```python
#用SN分群
total = purchase_data.groupby('SN').sum()['Price'].round(2)
average = purchase_data.groupby('SN').mean()['Price'].round(2)
count =purchase_data.groupby('SN').count()['Price']

summary = pd.DataFrame({'Purchase Count':count,
             'Average Purchase Price':average,
             'Total Purchase Value':total})
#根據Total Purchase，由大到小排列
summary = summary.sort_values(by='Total Purchase Value',ascending=False)

#加$符號
summary['Total Purchase Value'] = summary['Total Purchase Value'].astype('str')
summary['Total Purchase Value'] = '$'+summary['Total Purchase Value']
summary['Average Purchase Price'] = summary['Average Purchase Price'].astype('str')
summary['Average Purchase Price'] = '$'+summary['Average Purchase Price']

summary.head()
```

# Most Popular Items

● Top 5 most popular items by purchase count:

    ■ Item ID
    
    ■ Item Name
    
    ■ Purchase Count
    
    ■ Item Price
    
    ■ Total Purchase Value

● Retrieve the Item ID, Item Name, and Item Price columns

● Group by Item ID and Item Name. Perform calculations to obtain purchase count, item,price, and total purchase value

● Create a summary data frame to hold the results

● Sort the purchase count column in descending order

● Display a preview of the summary data frame
ss Crusade

```python
#用Item ID & Item name 分類
count = purchase_data.groupby(['Item ID','Item Name']).count()['Purchase ID']
total = purchase_data.groupby(['Item ID','Item Name']).sum()['Price'].round(2)
price = purchase_data.groupby(['Item ID','Item Name']).mean()['Price'].round(2)

summary = pd.DataFrame({'Purchase Count':count,
                       'Item Price':price,
                       'Total Purchase Value':total})
#根據Purchase Id，由大到小排列
summary= summary.sort_values(by='Purchase Count',ascending=False)

#加$符號
summary['Total Purchase Value'] = summary['Total Purchase Value'].astype('str')
summary['Total Purchase Value'] = '$'+summary['Total Purchase Value']
summary['Item Price'] = summary['Item Price'].astype('str')
summary['Item Price'] = '$'+summary['Item Price']

summary.head()
```
